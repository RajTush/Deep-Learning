```python
X_train = X_train.astype("float")/255.0
X_test = X_test.astype("float")/255.0
```

```python
plt.matshow(X_train[59999])
```

```python
X_train_falttened = X_train.reshape(len(X_train),28*28)
```

```python
X_test_falttened = X_test.reshape(len(X_test),28*28)
X_test_falttened.shape
```

```python
X_train_falttened.shape
```

We reshaped the X_train values becouse we have to convert the two dimentional array into a single dimensional array. Also Multi-Dimensional arrays take more amount of memory while 1-D arrays take less memory, which is the most important reason why we flatten the Image Array before processing/feeding the information to our model

```python
model = keras.Sequential([keras.layers.Dense(10,input_shape = (784,),activation = 'sigmoid')])

model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
model.fit(X_train_falttened, y_train, epochs = 10 )
```

```python
model.evaluate(X_test_falttened,y_test)
```

```python
plt.matshow(X_test[45])
```

```python
y_predicted = model.predict(X_test_falttened)
y_predicted[45]
```

```python
np.argmax(y_predicted[45])
```

```python
y_pred = [np.argmax(i) for i in y_predicted]
```

```python
# import the metrics class
from sklearn.metrics import confusion_matrix 
cnf_matrix = tf.math.confusion_matrix(labels= y_test,predictions = y_pred)
cnf_matrix
```

```python
import seaborn as sns
plt.figure(figsize = (10,10))
sns.heatmap(cnf_matrix,annot = True,fmt = 'd')
plt.xlabel('predicted')
plt.ylabel('Actual')
```

```python
modell = keras.Sequential([
            keras.layers.Dense(100,input_shape = (784,),activation = 'relu'),
            keras.layers.Dense(10,activation = 'sigmoid')
            ])

modell.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
modell.fit(X_train_falttened, y_train, epochs = 10 )
```

```python
modell.evaluate(X_test_falttened,y_test)
```

```python
plt.matshow(X_test[45])
```

```python
y_predictedd = modell.predict(X_test_falttened)
y_predictedd[45]
```

```python
np.argmax(y_predictedd[45])
```

```python
y_pred = [np.argmax(i) for i in y_predictedd]
```

```python
# import the metrics class
from sklearn.metrics import confusion_matrix 
cnf_matrixx = tf.math.confusion_matrix(labels= y_test,predictions = y_pred)
cnf_matrixx
```

```python
import seaborn as sns
plt.figure(figsize = (10,10))
sns.heatmap(cnf_matrixx,annot = True,fmt = 'd')
plt.xlabel('predicted')
plt.ylabel('Actual')
```

```python
#Selecting Dependent and independent variables
X = df.drop('PRICE',axis = 1)
y = df.PRICE
```

```python
from sklearn.model_selection import train_test_split
```

```python
# spliting data into train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 20)
```

```python
# Standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_train_scaled.shape
```

```python
X_test_scaled.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# importing tensorflow library for neural network analysis
from tensorflow import keras
from sklearn.model_selection import cross_val_score
```

```python
# Building Neural Network Model
model = keras.Sequential(
    [keras.layers.Dense(128,input_shape = (13,),activation = 'leaky_relu'),
     keras.layers.Dense(64,activation = 'leaky_relu'),
     keras.layers.Dense(1,activation = 'linear')])

model.compile(optimizer = keras.optimizers.SGD(0.001,0.9),
              loss = 'mean_squared_error',
              metrics = ['mae'])

history = model.fit(X_train_scaled,y_train,validation_split = 0.2,epochs = 100)
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1,len(loss)+1)
plt.plot(epochs,loss,'y',label = 'Training_loss')
plt.plot(epochs,val_loss,'r',label = 'validation_loss')
plt.legend()
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
loss_mae = history.history['mae']
val_loss_mae = history.history['val_mae']
epochs = range(1,len(loss)+1)
plt.plot(epochs,loss_mae,'y',label = 'Training_loss_for_mae')
plt.plot(epochs,val_loss_mae,'r',label = 'validation_loss_for_mae')
plt.legend()
plt.title('Training and Validation loss for MAE')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
prediction = model.predict(X_train_scaled[:5])
print(f'predicted values are : {prediction}')
print(f'real values are : {y_train[:5]}')
```

```python
mse_neural,mae_neural = model.evaluate(X_test_scaled,y_test)
print(f'MSE from neural_network is : {mse_neural}')
print(f'MAE from neural_network is : {mae_neural}')
```

Decision Tree

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
dtr  = DecisionTreeRegressor(random_state = 20)
dtr.fit(X_train_scaled,y_train)
y_train_pred = dtr.predict(X_train_scaled)
y_test_pred = dtr.predict(X_test_scaled)
mse_rf = mean_squared_error(y_test,y_test_pred)
mae_rf = mean_absolute_error(y_test,y_test_pred)
print(f'mean squared error for Decision Tree is : {mse_rf}')
print(f'mean absolute error for Decision Tree is : {mae_rf}')
```

```python
feature_list = list(X.columns)
feature_imp = pd.Series(dtr.feature_importances_, index = feature_list).sort_values()
print(feature_imp)
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error
rfr = RandomForestRegressor(n_estimators = 30,random_state = 20)
rfr.fit(X_train_scaled,y_train)
test = rfr.predict(X_test_scaled)
mse_rf = mean_squared_error(y_test,test)
mae_rf = mean_absolute_error(y_test,test)
print(f'mean squared error for Random forest is : {mse_rf}')
print(f'mean absolute error for Random forest is : {mae_rf}')
```

```python
feature_list = list(X.columns)
feature_imp = pd.Series(rfr.feature_importances_, index = feature_list).sort_values()
print(feature_imp)
```

From this we can say that the Normal Machine Learning Algorithms work better than the neural network model in this type of predicting Regression based dataset.

The Neural Network model giving Mean Absolute error Value of 2.81 And The Random forest model giving the Mean Absolute Error value of 2.8045. if we increase the no. of n_estimators in random forest the value of mae will ultimately decrease. so it is fair to say that the random forest is best model for this type of dataset

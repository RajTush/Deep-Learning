```python
# Importing Necessory Libraries
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as pd
import matplotlib.pyplot as plt
```

```python
# Loading the Fashion MNIST dataset
(X_train,y_train),(X_test,y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
plt.matshow(X_train[0])
```

```python
plt.figure(figsize = (4,4))
for i in range(15):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.matshow(X_train[i],cmap = plt.cm.binary)
  plt.xlabel(class_name[y_train[i]])
  plt.show()
```

```python
X_train_falttened = X_train.reshape(len(X_train),28*28)
X_train_falttened.shape
```

```python
X_test_falttened = X_test.reshape(len(X_test),28*28)
X_test_falttened.shape
```

```python
X_train_falttened[0]
```

```python
model = keras.Sequential([keras.layers.Dense(10,input_shape = (784,),activation = 'softmax')])

model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
model.fit(X_train_falttened, y_train, epochs = 150 )
```

```python
model.evaluate(X_test_falttened,y_test)
```

```python
plt.matshow(X_test[45])
```

```python
y_predicted = model.predict(X_test_falttened)
y_predicted[45]
```

```python
import numpy as np
np.argmax(y_predicted[45])
```

```python
plt.matshow(X_test[99])
```

```python
y_predicted[99]
```

```python
np.argmax(y_predicted[99])
```

```python
y_pred = [np.argmax(i) for i in y_predicted]
```

```python
# import the metrics class
from sklearn.metrics import confusion_matrix 
cnf_matrix = tf.math.confusion_matrix(labels= y_test,predictions = y_pred)
cnf_matrix
```

```python
import seaborn as sns
plt.figure(figsize = (10,10))
sns.heatmap(cnf_matrix,annot = True,fmt = 'd')
plt.xlabel('predicted')
plt.ylabel('Actual')
```

```python
model1 = keras.Sequential([keras.layers.Dense(128,input_shape = (784,),activation = 'relu'),
                          keras.layers.Dense(64,activation = 'relu'),
                          keras.layers.Dense(10,activation = 'softmax')])

model1.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
model1.fit(X_train_falttened, y_train, epochs = 50 )
```

```python
model1.evaluate(X_test_falttened,y_test)
```

```python
y_predictedd = model1.predict(X_test_falttened)
y_predictedd[567]
```

```python
plt.matshow(X_test[567])
```

```python
np.argmax(y_predictedd[567])
```

```python
y_predd = [np.argmax(i) for i in y_predictedd]
```

```python
# import the metrics class
from sklearn.metrics import confusion_matrix 
cnf_matrix = tf.math.confusion_matrix(labels= y_test,predictions = y_predd)
cnf_matrix
```

```python
import seaborn as sns
plt.figure(figsize = (10,10))
sns.heatmap(cnf_matrix,annot = True,fmt = 'd')
plt.xlabel('predicted')
plt.ylabel('Actual')
```
